1. What is the total AI revenue generated by each company?

```python
#1
v=a.groupby("Company")["AI_Revenue_USD_Mn"].sum()
v
```

```python
import matplotlib.pyplot as plt
plt.pie(v.values,autopct='%1.1f%%',labels=v.index,startangle=90)
plt.show()
```

1.google is the higest AI revenue generated   
2.open AI is generate the lowest revenue   
3.meta has 39.2 % of revenue generated

```python
#2
x=a.groupby("Company")["R&D_Spending_USD_Mn"].mean()
x
```

```python
import matplotlib.pyplot as plt
plt.pie(x.values,autopct='%1.1f%%',colors=["red","blue","yellow"],labels=x.index,startangle=90)
plt.show()
```

1.google is most spending on R&D spending_USD_mn     
2.open AI is least spending on R&D spending_USD_mn     
3.meta is second highest spending(39.2%)

```python
#3
y=a.groupby("Company")["AI_Revenue_Growth_%"].mean()
y
```

```python
import matplotlib.pyplot as plt
plt.pie(y.values,autopct='%1.11f%%',colors=["red","blue","yellow"],labels=y.index,startangle=90)
plt.show()
```

1.google has the higest average of AI revenue growth     
2.meta  has the  second higest average of AI revenue growth    
3.open has the lowest average of AI revenue growth

```python
#4
c=a.groupby("Company")["Stock_Impact_%"].sum()
c
```

```python
import matplotlib.pyplot as plt
plt.pie(c.values,autopct='%1.1f%%',colors=["red","blue","yellow"],labels=c.index,startangle=90)
plt.show()
```

1.Open AI has the higest stock impact            
2.meta has the lowest stock impact %     
3.google has the second  higest stock impact %

```python
#5
a["Year"] = a["Date"].dt.year
p=a.groupby("Year")["AI_Revenue_USD_Mn"].mean()
p
```

```python
plt.figure(figsize=(8,6))
plt.bar(p.index,p.values,color="blue",label="year")
plt.title("AI revenues per year")
plt.xlabel("years")
plt.ylabel("revenue")
plt.grid()
plt.show()
```

1. 2024 has the higest AI revenue per year    
2. 2015 has the lowest AI revenue per year    
3.2016 has the second lowest AI revenue per year    

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 2, figsize=(12, 10))

axs[2,1].bar(p.index, p.values, color="blue")
axs[2,1].set_title("AI revenues per year")

axs[0,1].pie(x.values, autopct='%1.1f%%', labels=x.index, startangle=90)
axs[0,1].set_title(" ")

axs[1,0].pie(y.values, autopct='%1.1f%%', labels=y.index, startangle=90)
axs[1,0].set_title("")

axs[1,1].pie(c.values, autopct='%1.1f%%', labels=c.index, startangle=90)
axs[1,1].set_title("")

axs[0,0].pie(v.values, autopct='%1.1f%%', labels=v.index, startangle=90)
axs[0,0].set_title("revenuue per company")

plt.tight_layout()
plt.show()
```
